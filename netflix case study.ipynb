# Replacing the missing values

```python
df['director'].fillna('Unknown Director',inplace = True)
df['cast'].fillna('Unknown cast',inplace = True)
df["country"].fillna("Unknown country",inplace=True)
df.dropna(subset=["date_added"],inplace=True)
df['rating'], df['duration'] = np.where(df['duration'].isna(), df['duration'], df['rating']), np.where(df['duration'].isna(), df['rating'], df['duration'])
df["rating"].fillna("TV-MA",inplace=True)
df.isnull().sum()

```

# Un-nesting the columns

Un-nest the column those have cell with  multiple commas seperated value by creating multiple multiple rows

```python
df_cast_r=pd.DataFrame(df["cast"].apply(lambda x:str(x).split(",")).tolist(),index=df["title"])
df_cast=df_cast_r.stack().reset_index()
df_cast.drop('level_1',axis=1,inplace=True)
df_cast.rename(columns={0:"cast"},inplace=True)
df_cast.head()
```

Un-nesting the Director column

```python
df_director_r = pd.DataFrame(df['director'].apply(lambda x: str(x).split(',')).tolist(), index =df['title'])

df_director = df_director_r.stack().reset_index()
df_director.drop('level_1', axis = 1, inplace = True)
df_director.rename(columns ={0:'director'}, inplace = True)
df_director.head()
```

merge both cast and director table

```python
df_new = df_director.merge(df_cast, how ='inner', on ='title')
df_new
```

```python
df_country_r=pd.DataFrame(df["country"].apply(lambda x:str(x).split(",")).tolist(),index=df["title"])
df_country=df_country_r.stack().reset_index()
df_country.drop("level_1",axis=1,inplace=True)
df_country.rename(columns={0:'country'},inplace=True)
df_country.head()
```

```python
df_listed_in_r=pd.DataFrame(df["listed_in"].apply(lambda x:str(x).split(",")).tolist(),index=df["title"])
df_listed_in=df_listed_in_r.stack().reset_index()
df_listed_in.drop("level_1",axis=1,inplace=True)
df_listed_in.rename(columns={0:'listed_in'},inplace=True)
df_listed_in.head()
```

```python
#Merge country column and listed_in
```

```python
df_new_1=df_country.merge(df_listed_in,how="inner",on="title")
df_new_1
```

```python
df_neww=df_new.merge(df_new_1,how="inner",on="title")
df_neww
```

```python
df.columns
```

#  Merge all columns after un-nesting

```python
df_final = df_neww.merge(df[['show_id', 'type', 'title', 'date_added',
       'release_year', 'rating', 'duration','description']] , how = 'inner', on = 'title')
df_final.sample(5)
```

# Data Visualisation

# How has the number of movies released per year changed over the last 20-30 years.

```python
movies_df=df_final[df_final['type'] == 'Movie']
current_year = pd.Timestamp.now().year
last_30_years_df = movies_df[movies_df['release_year'] >= (current_year - 30)]

movie_per_year=last_30_years_df.groupby("release_year")["type"].count()

```

```python
plt.plot(movie_per_year,color="r",marker="o")
plt.xlabel("movie_per_year")
plt.ylabel("Number of movies")
plt.title("Number of movies released (last 30 years)")
plt.grid()
plt.show()
```

In 2019 most movies are produced compare to other year.1995-2019 number of movies gradually increases.

# Does netflix has more focus on Tv show than movies in recent years

```python
year_count = df.groupby(["year_added" , "type"]).size().reset_index()
year_count.rename({0:"count"},axis=1,inplace=True)
year_count
```

```python
plt.figure(figsize = (20,5))
sns.barplot(data = year_count , x = "year_added" , y = "count" , hue = "type")
plt.legend(ncol = 2)
plt.show()
```

It is seen that from 2014 till 2021 the trend of movies and Tv shows have been increased. There are lot of movies and Tv shows are added into the netflix platform. However, the trend of graph of movie is little bit  decreasing after 2019 and TV show graph is increases compare of previous year. The recommendation increase the number of movies will eventually increase the users.

#According to graph best month to launch TV Show july ,sep and dec

```python
week = df.groupby('week_added')['type'].value_counts()
week = week.reset_index()
week['week_added'] = week['week_added'].astype('int64')
week
```

```python
week_TVShow = week.loc[week['type'] == 'TV Show'].sort_values(by = 'week_added')
week_TVShow
```

```python
sns.lineplot(data = week_TVShow, x = 'week_added', y = 'count',marker = 'o')
plt.xlabel("week_added")
plt.ylabel("count of TV_Show weekly")
plt.title("Best week to launch Tv Show")
plt.show()
```

The highest number of TV shows tends to be added in the 26th week(mostly in july)
